**Converting RGB image to grayscale image using openCV**

```python
# Importing open cv library
import cv2
```

```python
img = cv2.imread("/content/Dog.jpg")
```

```python
type(img)
```

```python
# Converting RGB image to grayscale image
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
print(grayscale_img.shape)
```

Now we can see that 3 is now gone because grayscale images only have 1 channel

cv2.imshow() will display the image. But, this will not work in Google colab.
